### Image binarization

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.color import rgb2gray
```

#### 1 Importation image

```python
image = imread('im12.jpg')[:,:,:3]
imshow(image)
```

#### 2 Conversion de l'image en nuances de gris

```python
fig, axis = plt.subplots(1, 1, figsize=(7,6))
image_gray = rgb2gray(image)
axis.imshow(image_gray, cmap='gray')
axis.set_title('Image en nuance de gris')
```

#### 3 Visualisation de la binarisation en fonction de chaque seuil

```python
seuils_test = np.linspace(0, 1, 13)

fig, axis = plt.subplots(3, 4, figsize=(15,8))

for seuil, ax in zip(seuils_test, axis.flatten()):
    image_binaire = image_gray < seuil
    ax.imshow(image_binaire)
    seuil = round(seuil,2)
    ax.set_title(f"SEUIL = {seuil}")
    
plt.tight_layout()
```

#### 4 Visualisation des images originale et binarise

```python
image_binaire = image_gray < 0.31

fig, axis = plt.subplots(1, 2, figsize=(15,8))

axis[0].imshow(image)
axis[0].set_title('Image originale')

axis[1].imshow(image_binaire)
axis[1].set_title("Image binarise")
```
